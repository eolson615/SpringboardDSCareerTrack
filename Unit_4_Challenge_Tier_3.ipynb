**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#I was perplexed by column 34, 37, and 47. Then I remembered the information above talking about the borough's only being 32. I realized #34-48 were not part of the London boroughs, and so they did not pertain to the question we were trying to answer.
#I used the .drop method, using .columns and the index number of the column method to specify which columns to drop.
properties = properties.drop(properties.columns[34:], axis=1)

#I then transposed the data as suggested
prop_transposed = properties.transpose()

#I then reset the index of the rows using the .reset_index method. This would allow the borough names to be a part of the data instead of the index names.
#I did feel that later on it would have been helpful to have the borough names as indexes. 
#Initially, out of habit I did the .head() to check if the transpose and the reset_index worked properly. In the next cell I check all of the rows to see if the .drop worked.
prop_index_reset = prop_transposed.reset_index()
prop_index_reset.head()
```

```python
#This was done to check to make sure the .drop was done correctly
prop_index_reset
```

```python
#I did this to rename the columns. I used .rename using columns= and the .iloc to specify that I wanted the first row, which was the DateTime entries to be my new column names. I then added the .drop to drop the first row with the DateTime data so I would not have it twice.
prop_fixedRowColumn = prop_index_reset.rename(columns=prop_index_reset.iloc[0]).drop(prop_index_reset.index[0])

#This is to see if it worked.
prop_fixedRowColumn.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#For clarities sake, I renamed the first two columns. Although I think at this point I probably could have just dropped the ID_Code column.
prop_fixedRowColumn = prop_fixedRowColumn.rename(columns={'Unnamed: 0':'Borough', pd.NaT:'ID_Code'})
#This was to make sure it worked
prop_fixedRowColumn.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#this was done to melt all of the Average Prices and the DateTimes into 2 columns. This is what was recommended, but I feel like it would have been easier to locate some of the price date keeping it this way. 
prop_melted = pd.melt(prop_fixedRowColumn, id_vars=['Borough', 'ID_Code'], var_name='Month', value_name='Average Price')
prop_melted.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#The Average_Price column had been an object, but I needed to convert it to a float so that I could do mathematical functions to it.
#To do this, I made a dictionary with the column name and the type I wanted all of the data in that column to be. Then I ran the dictionary in the .astype() method to convert it.
convert_dict = {'Average Price': float}
prop_melted = prop_melted.astype(convert_dict)
prop_melted.dtypes
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
#This was how I came up with the information for my plot. Like I mentioned earlier, I think the location notation could have been easier if the dataframe had not been melted earlier.
#I realized that I should change the column name of Average Price to Average_Price so I could use the df.columnName notation for location more easily.
prop_melted = prop_melted.rename(columns= {'Average Price':'Average_Price'})

#To help me from becoming confused, I used variables x and y to designate what data I was tryiing to place on the x axis and what data I was trying to place on the y axis.
#for the x axis I searched the Borough column for the name Brent, and then accessed the Month Date time entry.
x = prop_melted.Month[prop_melted.Borough == 'Brent']

#For the y axis I did the same as above, except instead of the date I accessed the price in the Average_Price column.
y = prop_melted.Average_Price[prop_melted.Borough == 'Brent']

#I then used plt.plot to make a simple line graph.
plt.plot(x, y)
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
#this was to check what the data types for each column were.
prop_melted.dtypes
```

```python
#This was the list of boroughs to iterate through in my for loop below. I probably could have just done, prop_melted[Borough] instead
Boroughs = ['City of London', 'Barking & Dagenham', 'Barnet', 'Bexley',
       'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield',
       'Greenwich', 'Hackney', 'Hammersmith & Fulham', 'Haringey',
       'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington',
       'Kensington & Chelsea', 'Kingston upon Thames', 'Lambeth',
       'Lewisham', 'Merton', 'Newham', 'Redbridge',
       'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets',
       'Waltham Forest', 'Wandsworth', 'Westminster']

#This was to create the dictionary from which I would build the new df with just two columns, the borough name and the change in the average price.
d_ofaverages = {}

#This is the for loop I used to obtain the average change in value
for borough in Boroughs:
  average1995 = prop_melted[(prop_melted.Borough == borough) & (prop_melted.Year == 1995)].mean()
  average2020 = prop_melted[(prop_melted.Borough == borough) & (prop_melted.Year == 2020)].mean()
  value = average2020.get('Average_Price') / average1995.get(key='Average_Price') 
  d_ofaverages[borough] = value

#This was to see the dictionary
d_ofaverages
```

```python
#This converted the dictionary into a df
df_averages = pd.DataFrame.from_dict(d_ofaverages, orient='index')
df_averages.head()
```

```python
#This was to reset the row index so that the boroughs could be part of the data and not an index again.
df_aver = df_averages.reset_index()
df_aver.head()
```

```python
#This was to rename the columns of the new df
df_aver = df_aver.rename(columns={'index':'Borough', 0:'AverageChange'})
df_aver.head()
```

```python
#This was to both order the df from highest change to lowest and to create a new column that showed the percentage change. This was for communication clarity. If I were presenting these findings, saying there was an 896% change is more easily understood than saying there was an 8.96 change in price.
df_aver_descending = df_aver.sort_values('AverageChange', ascending=False)
df_aver_descending['Perc_Change'] = df_aver_descending['AverageChange'] * 100
df_aver_descending
```

```python
#I did this to double check that Hackney really did have this large of a housing price increase.
x = prop_melted.Month[prop_melted.Borough == 'Hackney']
y = prop_melted.Average_Price[prop_melted.Borough == 'Hackney']
plt.plot(x, y)
```
