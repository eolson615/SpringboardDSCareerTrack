###Importing data from my GitHub into a pandas data frame.

###I notice that customerID is mixed alpha-numeric. There appear to be numerous categorical features. I will need to see the different unique categorical entries there are for those columns and how many there are of each type of categorical entry there are. The SeniorCitizen column appears to already be 1 and 0's so I will be expecting that to be a numeric column. tenure, MonthlyCharges, and TotalCharges appear to also be numeric so if they are not then that is an indication that there might be missing values or anomalous entries. The target feature is the Churn column. It appears to be Yes and No, with the Yes category being the one we are interested in for this business problem. It appears this column should be an object or string dtype.

```
url_df = 'https://raw.githubusercontent.com/eolson615/SpringboardDSCareerTrack/master/Capstone2/Data/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(url_df)
df.head()
```

###Below, I take a look at the column names to see if there are any strange spaces or naming practices. Then I will look at the unique categories in each column that has an object dtype. ###I need to address TotalCharges. It should be a float dtype, but it is currently an object dtype. I convert it to a float using the to_numeric method. This will allow any entries which are not numbers to be converted into NaN's. 

###I then study the rows that now have NaN's. I notice that all of the rows that have NaN's have a tenure = 0. I look at all the rows where tenure = 0. It turns out that the two sets of rows: those with NaN's and those with tenure = 0 are the same. So I convert the NaN's to 0, as these are brand new customers and they do not have any TotalCharges yet. I then check to see if there are anymore NaN's in the data frame.

```
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dtypes
```

```
df['TotalCharges'].isnull().sum()
```

```
nanrows = df[df.isna().any(axis=1)]
nanrows
```

```
tenureis0 = df[df['tenure'] == 0]
tenureis0
```

```
df['TotalCharges'] = df['TotalCharges'].replace(np.nan, 0)
df[df.isna().any(axis=1)]
```
